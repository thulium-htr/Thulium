## 2. Supported Languages

Thulium supports 52+ languages across 10 scripts.

````python
from thulium.data.language_profiles import list_supported_languages

languages = list_supported_languages()
print(f"Supported languages: {len(languages)}")
print(f"Examples: {languages[:10]}")
````

## 3. Language Details

Get detailed information about a language profile:

````python
from thulium.data.language_profiles import get_language_profile

# German language profile
profile = get_language_profile("de")
print(f"Name: {profile.name}")
print(f"Script: {profile.script}")
print(f"Region: {profile.region}")
````
